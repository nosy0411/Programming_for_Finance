```python
# 필요한 패키지들을 임포트한다.
# 웹 요청하는 모듈은 import requests 도 있지만 urllib도 있다. 기본적으로 파이썬에서 제공하는 표준라이브러리이다.
# urllib는 request보다 사용법이 어렵고 더 고수준 HTTP 클라이언트 인터페이스로 Requests 패키지를 권장하지만 
# 여기서는 표준 라이브러리인 urllib을 한번 사용해본다.

from urllib.request import urlopen
import pandas as pd
from bs4 import BeautifulSoup
import datetime
import json
import os
 

# 한국은행API 인증코드. 본인 아이디의 인증코드를 입력한다.
# 개인 인증키라 공개 깃허브에 올릴 수 없어 약간의 처리를 해주었다.

# secrets.json 가 있는 파일 경로를 설정한다.
path = os.getcwd()

# working directory를 변경
os.chdir(path)

# path = r'C:\Users\inventor\Desktop\curi\Curriculum_for_FR_Coding\2. Crawling\6주차'
secret_file = os.path.join(path, 'secrets.json') # secrets.json 파일 위치를 명시

with open(secret_file) as f:
    secrets = json.loads(f.read())

def get_secret(setting, secrets=secrets):
    """비밀 변수를 가져오거나 명시적 예외를 반환한다."""
    try:
        return secrets[setting]
    except KeyError:
        print(f"Set the {setting} environment variable")


# key에 회원가입 및 open api에서 신청한 secret key를 입력한다.
# 이 자료를 읽는 다른 분들은 위에 모듈 import 한거빼고 다 주석처리하고 key="개인비밀키" 하면된다.
key = get_secret("SECRET_KEY")
# print(key)

# 즉 , key = '개인키' 로 설정하면 된다.
```

```python
# https://ecos.bok.or.kr/jsp/openapi/OpenApiController.jsp?t=guideServiceDtl&apiCode=OA-1040&menuGroup=MENU000004
# 여기서 주소 만들면 쉽다. (통계 조회 조건 설정)
url="http://ecos.bok.or.kr/api/StatisticSearch/"+key+"/json/kr/1/10000/060Y001/DD/190001/202008/010200000/?/?/"

# 해석을 해보면 http://ecos.bok.or.kr/api/ + 서비스명(통계조회조건설정) StatisticSearch + 인증키 + 요청타입(json) + 언어(kr) + 요청시작건수(1) + 요청종료건수(10000) + 통계코드(4.1.1 시장금리(일별) : 060Y001) + 주기(일별 DD) + 검색시작일자(1900.01) + 검색종료일자(2020.08) + 항목코드1 (국고채 3년 : 010200000) + 항목코드2 + 항목코드3

result = urlopen(url)
html = result.read()
data = json.loads(html)

print(data)
```

```python
pd.set_option('display.max_columns', 12)  
data_row=data["StatisticSearch"]["row"]
print(data_row)
```

```python
pd.set_option('display.max_columns', 12)  

Interest_rate = pd.DataFrame(data_row)
Interest_rate_indexing = Interest_rate[Interest_rate["ITEM_CODE1"]=="010200000"]

# 그래프 y축값에 금리를 그려야 하는데 string 형으로 되어있으면 그림을 못그리므로 float형으로 바꿔준다.
Interest_rate_indexing["DATA_VALUE"]=Interest_rate_indexing["DATA_VALUE"].astype("float")

# 그래프 x축 값은 날짜를 나타내야 하는데 string 형이 아닌 datetime 형으로 바꿔준다.
Interest_rate_indexing["TIME"] = Interest_rate_indexing["TIME"].apply(lambda x: datetime.datetime.strptime(x, '%Y%m%d'))
print(Interest_rate_indexing)
```

```python
# 날짜별로 정렬해주고 데이터프레임의 인덱스를 TIME 열로 설정
Interest_rate_indexing=Interest_rate_indexing.sort_values("TIME")
Interest_rate_indexing.set_index("TIME",inplace=True)
print(Interest_rate_indexing)
```

# Plotly 시각화 패키지

- 사이트 https://plotly.com/python/

![plotly](./plotly.png)

- matplotlib, seaborn 보다 훨씬 세련된 동적인 그래프를 그릴 수 있다.

- d3.js를 이용하여 보다 interactive하게 그래프를 만들어준다. (D3.js는 웹브라우저 상에서 동적이고 인터렉티브한 정보시각화를 구현하기 위한 자바스크립트 라이브러리)

- 동적으로 확대, 확장, 이동, 회전 가능하며, 마우스를 대면 값 표시도 된다.

- 응용은 여기 참고 : https://www.kaggle.com/hakkisimsek/plotly-tutorial-3



```python
# 임포트
import plotly.graph_objs as go

# 오프라인 모드를 사용한다. 온라인모드는 유료다.
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

# 주피터 노트북이나 파이썬 개발환경 내에서 그래프를 보여주는 것을 선택한다.
init_notebook_mode(connected=True)

# Plotting : 2차원 데이터 분포를 그리기 위해 go.Scatter()를 사용한다.
# matplotlib과 마찬가지로 x,y에는 iterable한 자료가 들어가면 된다.
# 이것을 최종적으로 trace라는 변수에 할당해준다.
trace = go.Scatter(x=Interest_rate_indexing.index, y=Interest_rate_indexing["DATA_VALUE"]) 

# go.Layout()을 사용해 그래프의 세부 디테일을 지정해줄 수 있다. matplotlib의 figure와 axes, axis를 변경하는 것과 같다.
layout = go.Layout( title='국고채(3년) 금리 추이', 
                    xaxis=dict(

                            # range selector 표시하기 
                            rangeselector=dict( 
                                buttons=list(
                                    [ dict(count=1, label='1m', step='month', stepmode='backward'), 
                                        dict(count=3, label='3m', step='month', stepmode='backward'), 
                                        dict(count=6, label='6m', step='month', stepmode='backward'), 
                                        dict(count=1, label="YTD", step="year", stepmode="todate"),
                                        dict(count=1, label="1y", step="year", stepmode="backward"),
                                        dict(count=2, label="2y", step="year", stepmode="backward"),
                                        dict(count=3, label="3y", step="year", stepmode="backward"),
                                        dict(count=5, label="5y", step="year", stepmode="backward"),
                                        dict(count=7, label="7y", step="year", stepmode="backward"),
                                        dict(count=10, label="10y", step="year", stepmode="backward"),
                                        dict(step='all') ]) 
                                    ), 
                            
                            # range slider 표시하기
                            rangeslider=dict(visible=True), 
                            type='date',
                            title='날짜'
                            ),
                        yaxis=dict(title='수익률'),
                        width=800, height=600
                        )

# range selector과 range slider에 대해 다양한 옵션은 여기 참고 : https://plotly.com/python/range-slider/

print(trace)
print(layout)
```

```python
# trace는 리스트에 넣는다. 만약에 한 화면에 그래프를 2개 그리고 싶으면 위에서 trace를 하나 더 만들어서
# data1 = [trace1, trace2] 와 같은 식으로 설정해주면 된다.
data1=[trace] 

# 다음과 같이 figure을 딕셔너리로 trace와 layout을 지정해준다.
fig1 = dict(data=data1, layout=layout)

# 그림그리기
iplot(fig1)
```
