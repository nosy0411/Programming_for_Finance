# 시계열 데이터 (Time Series)

- 시계열 데이터는 관측 값들이 시간에 따라 변하는 자료를 말한다 . (ex : 주가 , 주가 수익률 , 변동성 등)

- 시계열 자료는 인접된 값들이 서로 완전히 독립적 이라고는 할 수 없고 , 상호 종속적 관계에 있다 . 현재는 과거에 종속된다 . 완전히 독립적 관계의 시계열 자료도 존재한다 . (ex : 백색 잡음 , 주가의 수익률)

- 시계열 자료는 일반적인 통계 이론이나 통계적 분석 방법으로는 다루기 어렵고 , 시계열 분석 (Time Series Analysis) 방법을 사용한다

- 대부분의 금융 관련 자료는 주식의 일별 시세 , 원 달러 시세 , 변동성 시세 등 ) 시계열 자료의 특성을 가진다

- 금융 시계열 자료는 통상의 시계열 자료 월별 출생자수 , 태양의 월별 흑점수 등 ) 보다 복잡한 구조를 갖는다


# 시계열 분석의 목적
- 시계열 자료의 자기종속 구조를 파악하고 이 종속 구조를 효과적으로 기술하는 모형 ( 을 개발하여 미래의 값이나 상황을 예측하는 것을 목적으로 한다

# 시계열 분석 방법

- 일반적인 시계열 자료는 그 종속 구조가 주로 시간에 따른 평균 일차 적률 에 의해 설명되는 반면 , 금융 시계열 자료의 종속 구조는 평균 일차 적률 은 물론이고 변동성을 설명하는 이차 적률 , 즉 시간에 따른 분산이나 표준편차들 사이의 종속구조까지 분석해야 한다

- 시계열 분석에서는 동일한 기간이나 같은 장소에서 독립적인 반복 관측이 허용되지 않고 기본적으로 시간에 따라 관측되는 한 계열자료만을 이용하기 때문에 반복 관측이 허용되는 분석 이론에 비해 더 일반화된 이론이 요구된다

- 시간영역 분석 : 인접한 시계열 자료 사이에 존재하는 종속성이나 상관성이 과거의 자료로 회귀한다는 가정에서 출발하여 시계열 자료를 추세와 같은 결정적 요인과 , 잡음과 같은 확률 성분으로 분해하여 분석한다 . (AR, MA, ARMA, ARIMA, SARIMA 등

- 주파수영역 분석 : 평균이나 분산 등이 비교적 안정된 정상 시계열을 대상으로 강한 주기함수들의 일차결합으로 분해하고 , 각 주기 함수로 시계열의 주기성을 파악한다 스펙트럼 분석 , Spectral

# 시계열 데이터의 특징

- 추세성 (Trend) : 자연 현상의 시계열 자료와 마찬가지로 금융 시계열 자료도 같은 방향으로 직선이나 곡선을 따라 상승하거나 하강하는 경향을 보인다 . 이것은 금융 현상에서 나타나는 성장 ( 이나 증가 등의 효과로 발생한다 . 추세는 크게 결정적 추세와 확률적 추세로 구분할 수 있다 . 결정적 추세는 시계열의 평균값이 시간의 함수로 주어지는 반면 , 확률적 추세는 특정 시구간에서 시계열의 성장이나 감소율이 과거 값들의 종속관계로 설명된다

- 계절성 (Seasonality) : 절기나 분기 또는 연 , 월 , 일별로 관측되는 금융 시계열은 대개 시간적 구분 사이에 존재하는 상관에 의한 계절적 패턴을 보인다 . 계절성의 특징은 넓은 의미에서 경제 규모와 현상의 확장과 수축의 순환과정에서 일어나는 비교적 긴 주기의 주기적 운동을 포함한다

- 이변성 (Aberration) : 금융 시계열 자료에서는 일부이긴 하지만 관측값들이 시간에 따라 진행되고 있는 일반적인 패턴이나 확률구조에서 크게 이탈하는 경향이 있다 . 이변성은 확률구조가 일정한 상태에서 발생한 단순한 이상값으로 보기는 어렵고 순간적이긴 하지만 근본적인 확률구조의 변화가 이 시점에서 발생한 것으로 본다 . 이변성은 금융 시계열의 예측모형 개발에서 예측 오차를 최소화하기 위해 반드시 고려해야 할 사항이다

- 변동성 (Volatility) : 시계열에서 분산이 시간의 추이에 따라 변하는 성질을 말하는 것으로 금융 시계열의 대표적인 특징이다 . 변동성은 새로운 정보의 유입 , 이에 대한 서로 다른 해석 , 외적 사건의 출현 상황에서 주어지는 시계열의 패턴이다 . 변동성은 관련 현상의 불확실성 및 위험 ( 을 설명하는 수단이 된다 . 변동성은 큰 값은 큰 값끼리 , 작은 값은 작은 값끼리 일정 기간 덩어리를 이루는 변동성 집중과 이분산의 형태로 나타난다

- 비선형성 (Nonlinearity) : 금융 시계열에서는 시간에 따라 변하는 값들 사이에 존재하는 종속 관계나 인과 관계가 자신의 값들 사이의 비선형관계로 설명되는 경우가 많다 . 비선형성은 정책이나 제도의 충격이 경제나 금융 현상에 영향을 주지만 그 영향의 정도가 반드시 그 충격의 강도에 비례하지는 않으며 , 또 충격의 강도가 같은 경우에도 그 영향은 서로 반대가 될 수도 있는 것과 관련이 있다

# 시계열 데이터의 정상성 

- 비 정상성 (Non stationary) : 비 정상성 시계열 자료는 추세 성분을 가지며 시간에 따라 확률구조가 변하는 시계열을 의미한다 . 즉 , 시간에 따라 평균 , 분산 같은 통계적 특성이 변한다 . 예를 들어 주가의 가격 데이터와 같이 랜덤워크 과정으로 구성된 시계열은 비 정상성을 보이며
분석이 어렵고 예측이 불가하다

- 정상성 (Stationary) : 정상성 시계열 자료는 시간이 변해도 확률구조가 일정한 시계열을 의미한다 . 시간이 지나도 평균 , 분산 왜도 , 첨도 등 모든 확률구조가 동일하면 강정상 (Strongly Stationary) 혹은 완전 정상 (Completely Stationary) 이라고 하며, 평균 , 분산 정도의 (2 차 적률 까지 ) 확률구조가 같으면 약정상 (Weakly Stationary) 이라 한다 . 강정상은 조건이 너무 까다로우므로 , 약정상 시계열을 보통 정상 시계열이라고 한다

- 참고 사항 ( 정상성에 대한 수학적 정의 ) : 시계열 자료에 대해 아래의 성질을 만족하면 (약) 정상 시계열이라고 한다  

    - (1) 평균이 모든 𝑡에서 일정 𝐸\[𝑋𝑡]=𝜇 평균이 상수  
    - (2) 자기공분산이 시차 (time lag)에만 의존함
      
        - 𝐶𝑥(𝑡,𝑠)=𝐶𝑜𝑣(𝑋𝑡,𝑋𝑠)=𝐸\[(𝑋𝑡–𝜇)(𝑋𝑠−𝜇)]=𝐶𝑥(𝑠−𝑡)

```python
import numpy as np
import pandas as pd
import math
from pandas.plotting import scatter_matrix
import pandas_datareader.data as web

import datetime as dt
import scipy.stats as stats

from statsmodels.tsa.arima_process import arma_generate_sample
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf

import matplotlib.pyplot as plt


# Yahoo site로 부터 주가 데이터를 수집
def getStockDataYahoo(stockCode, start='', end=''):
    # 수집 기간
    if start == '':
        start = dt.datetime(2007, 1, 1)
    else:
        start = dt.datetime.strptime(start, '%Y-%m-%d')

    if end == '':
        end = dt.date.today()
    else:
        end = dt.datetime.strptime(end, '%Y-%m-%d')

    stock = pd.DataFrame()

    # Yahoo site로 부터 주가 데이터를 수집한다. 가끔 안들어올 때가 있어서 10번 시도한다.
    for i in range(0, 10):
        try:
            # 수정 주가로 환산하여 읽어온다
            stock = web.YahooDailyReader(
                stockCode, start, end, adjust_price=True).read()
        except:

            print("%s not collected (%d)" % (stockCode, i + 1))

        if not stock.empty:
            break

    if stock.empty:
        print("%s not collected" % stockCode)

    # Volume이 0 인 경우가 있으므로, 이를 제거한다
    stock = stock.drop(stock[stock.Volume < 10].index)

    # 데이터에 NA 값이 있으면 제거한다
    stock = stock.dropna()

    # 수집한 데이터를 파일에 저장한다.
    stock.to_csv('StockData/' + stockCode[0:6] + '.csv', date_format='%Y-%m-%d')
    print("%s 데이터를 수집하였습니다. (rows = %d)" % (stockCode, len(stock)))
    return stock

# 지수이동평균을 계산한다
# data : Series
def EMA(data, n):
    ma = []
    
    # data 첫 부분에 na 가 있으면 skip한다
    x = 0
    while True:
        if math.isnan(data[x]):
            ma.append(data[x])
        else:
            break
        x += 1
        
    # x ~ n - 1 기간까지는 na를 assign 한다
    for i in range(x, x + n - 1):
        ma.append(np.nan)
    
    # x + n - 1 기간은 x ~ x + n - 1 까지의 평균을 적용한다
    sma = np.mean(data[x:(x + n)])
    ma.append(sma)
    
    # x + n 기간 부터는 EMA를 적용한다
    k = 2 / (n + 1)
    
    for i in range(x + n, len(data)):
        #print(i, data[i])
        ma.append(ma[-1] + k * (data[i] - ma[-1]))
    
    return pd.Series(ma, index=data.index)

# MACD 지표를 계산한다
# MACD Line : 12-day EMA - 26-day EMA
# Signal Line : 9-day EMA of MACD line
# MACD oscilator : MACD Line - Signal Line
# ----------------------------------------
def MACD(ohlc, nFast=12, nSlow=26, nSig=9, percent=True):
    ema1 = EMA(ohlc.Close, nFast)
    ema2 = EMA(ohlc.Close, nSlow)
    
    if percent:
        macdLine =  100 * (ema1 - ema2) / ema2
    else:
        macdLine =  ema1 - ema2
    signalLine = EMA(macdLine, nSig)
    
    return pd.DataFrame(macdLine - signalLine, index=ohlc.index)
```

```python
# Yahoo 사이트에서 주가 데이터를 수집하여 주가, 거래량, 수익률, MACD 지표를
# 관찰하고, 비정상 시계열 (Non-stationary)과 정상 시계열 (Stationary)의 차이점을 관찰한다.
# ----------------------------------------------------------------------

df = getStockDataYahoo('102110.KS', start='2010-01-01')
df = pd.read_csv('StockData/102110.csv', index_col=0, parse_dates=True)

# 종가를 기준으로 일일 수익률을 계산한다.
df['Rtn'] = np.log(df['Close']) - np.log(df['Close'].shift(1))

# MACD 기술적 지표를 측정한다
df['macd'] = MACD(df)
df = df.dropna()

# 주가, 거래량, 수익률, MACD를 그린다
fig = plt.figure(figsize=(10, 6))
p1 = fig.add_subplot(2, 2, 1)
p2 = fig.add_subplot(2, 2, 2)
p3 = fig.add_subplot(2, 2, 3)
p4 = fig.add_subplot(2, 2, 4)

p1.plot(df['Close'], color='blue', linewidth=1)
p2.plot(df['Volume'], color='red', linewidth=1)
p3.plot(df['Rtn'], color='purple', linewidth=1)
p4.plot(df['macd'], color='green', linewidth=1)
p1.set_title("Stock Price")
p2.set_title("Volume")
p3.set_title("Return")
p4.set_title("MACD oscilator")
p3.axhline(y=0, color='black', linewidth=1)
p4.axhline(y=0, color='black', linewidth=1)
plt.tight_layout()
plt.show()
```

# 시계열 모형 시간 영역 분석

- AR 모형 : 평균회귀 모형 (Auto Regressive Model) 으로 현재 시계열은 과거 시계열 데이터에 종속된다는 모형이다 . 과거와의 종속성이 거의 없으면 현재 데이터는 과거와는 독립적이며 백색 잡음 (white noise) 시계열이 된다 (ex : 주가의 수익률 ). 과거와의 종속성이 강할수록 현재 데이터는 과거에 의존적이며 랜덤워크에 가까워 진다 (ex : 주가 ). AR 모형은 과거와의 자기상관성을 분석하여 대상 시계열의 특성을 분석한다

- MA 모형 : 이동평균 과정 (Moving Average Model) 으로 현재 시계열 데이터는 과거 잔차들의 가중 평균으로 구성된다는 모형이다 . 잔차항은 백색 잡음이므로 현재 데이터는 과거의 백색 잡음들의 평균 같은 값으로 설명된다는 모형이다 . 백색 잡음들은 정상성 (Stationary) 이 높고 평균회귀 특성이 있으므로 이 들의 합으로 구성된 MA 모형도 평균회귀 특성이 있다

- ARMA 모형 : AR 모형과 MA 모형을 혼합한 모형이다 . 현재 시계열은 과거 시계열 데이터들과 과거 잔차들의 조합으로 설명된다는 모형이다 . AR 모형과 MA 모형이 모두 평균회귀 특성이 있으므로 이 들의 조합인 ARMA 모형도 평균회귀 특성이 있다 . AR, MA, ARMA 모형은 평균회귀 특성이 있으므로 정상성(Stationary) 을 갖는 시계열 분석에 적합하다 (ex : 주가의 기술적분석지표 , 페어 트레이딩의 스프레드 시계열 분석 등

- ARIMA 모형 : 시계열 자체로는 정상성이나 평균회귀 특성이 없어도 이를 차분한 (Indifference) 시계열은 평균회귀 특성이 있을 수 있다 (ex : 주가를 차분하면 주가 수익 시계열이 됨 ). ARIMA 모형은 차분 시계열에 ARMA 모형을 적용한 모형이다 . 주가는 랜덤워크 특성이 나타나지만 주가의 차분 시계열은 정상성을 갖는 시계열이 되므로 ARIMA 모형을 적용해 볼 수 있다 . 차분 값 (Indifference) = 0 인 경우는 ARMA 모형이 됨

- SARIMA 모형 : 주기성 (Seasonality) 을 갖는 시계열 모형이다 . ARMA 모형이나 ARIMA 모형이 주기성을 갖는 경우 , 자기상관성을 분석하면 주기를 검출할 수 있다 . 예 : 일 중 (Intraday) 시간대 별 거래량을 관찰하면 장 초반에는 거래량이 높고 , 중반에는 거래량이 낮다가 , 후반에는 다시 거래량이 높아진다 (U-shaped). 30 일 정도의 시계열을 관찰하면 U 자 모양이 주기적으로 반복해서 나타나고 주기는 1 일이 된다 . 장 중 변동성도 유사한 모습임

# AR 모형 : 자기회귀 모형

- AR 모형은 평균으로 회귀하려는 경향을 갖는 시계열로 , 이전 시점의 자신의 값과 현재 시점의 잡음 성분의 선형 조합으로 표현되는 시계열이다

    - 𝐴𝑅(𝑝) ∶ 𝑋𝑡 = 𝑎1𝑋𝑡−1 + 𝑎2𝑋𝑡−2 + 𝑎3𝑋𝑡−3 + ⋯ + 𝜀𝑡 ← 𝑒𝑥) 𝑋𝑡 : 금일 주가 수익률 , 𝑋𝑡−1 ∶ 어제 주가 수익률 , 𝑎𝑖 ∶ 현재와 과거의 관련 정도  

- 𝐴𝑅1 ∶ 𝑋𝑡 = 𝑎𝑋𝑡−1 + 𝜀𝑡 ← 𝐴𝑅(𝑝)모형을 단순화 시킨 것으로 금일 데이터는 어제 데이터에만 종속되는 모형

- 𝐴𝑅(1)모형에서 |𝑎|<1이면 시계열 𝑋는 정상성 (Stationary) 을 갖는 시계열이 되고 , 아니면 랜덤워크가 된다

- 𝐴𝑅(1)모형에서 𝑋가 정상성을 갖으면 𝑎는 𝑋의 자기상관계수 (시차 = 1) 와 일치한다

- 𝑎가 0에 가까울수록 𝑋의 정상성은 강해지고 ( 평균으로 회귀하는 속도가 빠르다 ), 𝑎가 1에 가까울수록 정상성이 약해져서 랜덤워크에 가까워진다

- 𝐴𝑅모형의 적용 예 : 주가의 수익률 , 페어 트레이딩의 스프레드 신호 등 정상성을 갖는 모든 시계열

- 𝐴𝑅(1)모형의 시계열은 𝑎값에 따라 아래와 같은 모습을 보인다

```python
# AR 모형을 임의로 생성하고, AR 파라메터가 변할 때 시계열의 모습이 어떻게
# 변하는지 육안으로 확인한다. AR 파라메터의 특성을 직관적으로 이해한다.
# 또한, AR 모형에 대한 ACF와 PACF 특성을 육안으로 확인한다.
# ACF/PACF는 향후 실제 시계열을 분석할 때 분석 모델을 선정에 참고한다.

from statsmodels.tsa.arima_process import arma_generate_sample
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf

# ARIMA(ar, d, ma) 모형으로 n개의 데이터를 샘플링한다
def sampleARIMA(ar, d, ma, n):
    arparams = np.array(ar)
    maparams = np.array(ma)
    ar = np.r_[1.0, -arparams]  # add zero-lag and negate
    ma = np.r_[1.0, maparams]  # add zero-lag

    # ARMA 모형으로 n개의 데이터를 샘플링한다
    y = arma_generate_sample(ar, ma, n)

    # 지정된 차분 횟수 (d) 만큼 누적한다
    for i in np.arange(d):
        y = np.cumsum(y)

    return y
```

```python
# --------------------------------------------------------------------

# AR(1) 모형의 a 값을 변화시켜 가면서 AR(1) 시계열을 생성하고 그 모습을 관찰한다
# a 값을 변화시켜 가면서 그래프의 모양이 어떻게 바뀌는지 확인한다.

y1 = sampleARIMA(ar=[0.1], d=0, ma=[0], n=500)
y2 = sampleARIMA(ar=[0.5], d=0, ma=[0], n=500)
y3 = sampleARIMA(ar=[0.9], d=0, ma=[0], n=500)
y4 = sampleARIMA(ar=[0.99], d=0, ma=[0], n=500)

fig = plt.figure(figsize=(10, 7))
p1 = fig.add_subplot(2,2,1)
p2 = fig.add_subplot(2,2,2)
p3 = fig.add_subplot(2,2,3)
p4 = fig.add_subplot(2,2,4)

p1.plot(y1, color='blue', linewidth=1)
p2.plot(y2, color='red', linewidth=1)
p3.plot(y3, color='purple', linewidth=1)
p4.plot(y4, color='green', linewidth=1)
p1.set_title("a = 0.1")
p2.set_title("a = 0.5")
p3.set_title("a = 0.9")
p4.set_title("a = 0.99")
plt.tight_layout()
plt.show()
```

```python
# a=0.5인 AR(1) 샘플을 생성하고, 거꾸로 이 샘플을 AR(1) 모형으로 분석하여 a값을 추정해본다.
y = sampleARIMA(ar=[0.5], d=0, ma=[0], n=500)
model = ARIMA(y, order=(1,0,0)).fit()
print(model.summary())
print('\n')

# AR(2) 모형도 확인해 본다
y = sampleARIMA(ar=[0.1, -0.4], d=0, ma=[0], n=500)
plt.plot(y, color='blue', linewidth=1)
model = ARIMA(y, order=(2,0,0)).fit()
print(model.summary())
```

```python
# AR 모형의 ACF와 PACF를 확인해 본다. a값을 변화시켜 가면서 비교해 본다
# ACF와 PACF는 향후 실제 시계열을 어느 모형으로 분석할 지에 대한 단서를 제공한다

# AR(1) 모형의 ACF (Autocorrelation) 는 지수적으로 감소하고 , PACF 는 급격히 감소하는 모습이다 -> AR 모형의 특징
# 분석 대상인 어떤 시계열 데이터의 ACF 특성이 아래와 유사하다면 그 시계열은 AR(1) 모형에 가까울 가능성이 있다 -> 모형을 선택하는 단서를 제공
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

y = sampleARIMA(ar=[0.5], d=0, ma=[0], n=500)
fig = plt.figure(figsize=(12, 3))
plt.plot(y, color='red', linewidth=1)
plt.show()

fig = plt.figure(figsize=(12, 4))
p1 = fig.add_subplot(1,2,1)
p2 = fig.add_subplot(1,2,2)
plot_acf(y, p1, lags=50)
plot_pacf(y, p2, lags=50)
plt.show()
```

```python
# 시계열 데이터의 정규성을 확인한다
def checkNormality(data):
    fig = plt.figure(figsize=(10, 8))
    p1 = fig.add_subplot(2, 2, 1)
    p2 = fig.add_subplot(2, 2, 2)
    p3 = fig.add_subplot(2, 2, 3)
    p4 = fig.add_subplot(2, 2, 4)

    p1.plot(data)  # 육안으로 백색 잡음 형태인지 확인한다
    p1.set_title("Data")

    # Residual의 분포를 육안으로 확인한다
    # residual는 정규분포라 가정하고 probability density function 그림
    r = np.copy(data)
    r.sort()
    pdf = stats.norm.pdf(r, np.mean(r), np.std(r))
    p2.plot(r, pdf)

    # Y축을 빈도수(frequency)가 아니라 density로 하고 싶을 때 density=True
    # 즉 y 값이 구간에서 확률 밀도 함수의 값이며 범위에 대한 적분이 1이 되도록 정규화 된다.
    # density=True와 normed=True는 같은 옵션이고, 두 개 옵션 모두 true로 두면 오류가 난다.
    p2.hist(r, density=True, bins=100)
    p2.set_title("Distribution")

    # Q-Q plot을 그린다
    stats.probplot(data, dist="norm", plot=p3)

    # ACF plot을 확인한다. 백색 잡음은 자기상관성이 없다.
    # 샘플 자기상관계수 함수와 표준 오차의 크기를 동시에 그려준다.
    # 참고 : https://datascienceschool.net/view-notebook/12c6c314797c4c438d06e5cf4d290840/
    plot_acf(data, lags=100, ax=p4)

    # Shapiro-Wilks 검정을 수행한다
    # (검정통계량, p-value)가 출력된다.
    # 귀무가설 : 정규분포 이다, 대립가설 : 아니다
    # p-value > 0.05 이면 귀무가설을 기각할 수 없다 --> "정규분포이다"
    w = stats.shapiro(data)
    print()
    print("Shapiro-Wilks 검정 : 검정통계량 = %.4f, p-value = %.4f" % (w[0], w[1]))
```

```python
# AR(2) 샘플을 생성한다
# a = (0.5, 0.5) 인 AR(2) 샘플 데이터를 다시 AR(2) 모델로 분석한 결과가 타당한지 확인해 본다 . 모형의 잔차가 정규분포 특성을 갖는지 확인한다

y = sampleARIMA(ar=[0.5, -0.5], d=0, ma=[0], n=1000)
plt.figure(figsize=(10, 6))
plt.plot(y, color='blue', linewidth=1)
plt.title("AR(2)")
plt.show()

# 샘플링 데이터를 AR(2) 모형으로 분석한다 (Fitting)
model = ARIMA(y, order=(2,0,0))
model_fit = model.fit()
print(model_fit.summary())

# Fitting이 잘되었는지 확인하기 위해 Residual을 분석한다.
# (1) Residual은 실제 데이터와 추정치의 차이이므로 백색 잡음 (잔차) 이어야 한다. (백색 잡음에 가까울 수록 좋음)
# (2) 따라서 Residual은 정규분포 특성을 가져야한다. (정규분포 모양에 가까울 수록 좋음)
# (3) 정규분포 특성을 조사하면  Fitting이 잘되었는지 확인할 수 있다. : Q-Q plot
# (4) 잔차 성분의 ACF : 자기상관관계가 작을 수록 좋다.
residual = model_fit.resid
checkNormality(residual)  # 육안으로 백색 잡음 형태인지 확인한다
```

```python

# a = 0.8 AR(1) 샘플을 생성한 후 향후 10 기간의 예측값을 추정한다 . 예측 결과가 의미하는 것은 무엇인가 ? 
# AR(1) 샘플을 생성한다
y = sampleARIMA(ar=[0.8], d=0, ma=[0], n=500)
plt.figure(figsize=(10, 3))
plt.plot(y, color='brown', linewidth=1)
plt.show()

# 샘플링 데이터를 AR(1) 모형으로 분석한다 (Fitting)
model = ARIMA(y, order=(1,0,0))
model_fit = model.fit()

# 향후 10 기간 데이터를 예측한다
forecast = model_fit.forecast(steps=10)[0]
forecast = np.r_[y[-1], forecast]  # y의 마지막 값을 forecast 앞 부분에 넣는다

# 원 시계열과 예측된 시계열을 그린다
ytail = y[len(y)-100:]   # 뒷 부분 100개만 그린다

ax1 = np.arange(1, len(ytail) + 1)
ax2 = np.arange(len(ytail), len(ytail) + len(forecast))

plt.figure(figsize=(10, 3.5))
plt.plot(ax1, ytail, color='blue', label='Time series', linewidth=1)
plt.plot(ax2, forecast, color='red', label='Forecast')
plt.axvline(x=ax1[-1],  linestyle='dashed', linewidth=1)
plt.title("Time Series Forcast")
plt.legend()
plt.show()
```

# 비정상 시계열 분석

- AR, MA, ARMA 모형은 정상 시계열 (Stationary) 분석에 적합하다

- 비정상 시계열은 정상 시계열로 변환하여 ARMA 모형을 적용하여 분석한 후 , 다시 비정상 시계열로 변환할 수 있다

- 비정상 시계열을 정상 시계열로 변환하는 방법은 차분 시계열로 변환하는 방법과 , 추세를 제거하는 방법 등을 생각해 볼 수 있다

- 차분 시계열은 정상 시계열이므로 ( 정상 시계열이 될 때까지 차분함 ) ARMA 모형을 적용할 수 있고 , 차분 시계열을 다시 누적시키면 원래의 시계열이 된다

- 비정상 시계열과 회귀직선같은 추세선의 차이는 정상 시계열이 되어 ARMA 모형을 적용할 수 있다 . 분석 후 다시 추세를 삽입하면 원래의 시계열이 된다

# ARIMA(p, d, q) 모형 : 샘플 데이터 생성
- ARIMA 모형은 비정상 시계열을 설명하려는 모형이다 . 비정상 시계열을 정상 시계열이 될 때까지 차분한 후 ARMA(p, q) 모형을 적용한다

- 차분 시계열은 원 시계열의 추세 성분만 제외되었을 뿐 , 원시계열의 종속성이나 변동성 등 다른 확률구조나 성질은 그대로 유지된다 . 추세 성분을 제외한 나머지 성질들은 ARMA 모형으로 분석이 가능하다

- 비정상 시계열이 구간에 따라 수준 ( 은 다를지라도 그 행태가 수준에 관계없이 유사성을 가지는 등질비정상 ) 시계열에 적합하다 . 금융 시계열은
등질비정상인가


```python
# ARIMA(1,1,1) 샘플을 생성한다
y = sampleARIMA(ar=[0.8], d=1, ma=[0.5], n=500)
d = np.diff(y) # 차분하면 ARMA(1,1)이 된다

# ARIMA 시계열과 차분 시계열을 그린다
fig = plt.figure(figsize=(12, 5))
p1 = fig.add_subplot(1,2,1)
p2 = fig.add_subplot(1,2,2)
p1.plot(y, color='blue', linewidth=1)
p2.plot(d, color='red', linewidth=1)
p1.set_title("ARIMA(1,1,1)")
p2.set_title("ARMA(1,1)")
plt.show()
```

# ARIMA 모형 : ARIMA(1,1,1) 모수 추정 및 성능 평가
- a = 0.8, b=0.5, 차분횟수 (d) = 1 인 ARIMA(1,1,1) 샘플 데이터를 생성하고 , 거꾸로 이 데이터를 ARIMA(1,1,1) 모형으로 분석하여 a, b 값을 추정해 본다

```python
# 샘플링 데이터를 ARIMA(1,1,1) 모형으로 분석한다 (Fitting)
y = sampleARIMA(ar=[0.8], d=1, ma=[0.5], n=500)
model = ARIMA(y, order=(1,1,1))
model_fit = model.fit()
print(model_fit.summary())
```

# ARIMA 모형 : ARIMA(1,1,1) 모수 추정 및 성능 평가
- a = 0.8, b=0.5, 차분횟수 (d) = 1 인 ARIMA(1,1,1) 샘플 데이터를 분석한 결과가 타당한지 확인한다 . 모형의 잔차가 정규분포 특성을 갖는지 확인한다

```python
# Fitting이 잘되었는지 확인하기 위해 Residual을 분석한다.
# (1) Residual은 실제 데이터와 추정치의 차이이므로 백색 잡음 (잔차) 이어야 한다. (백색 잡음에 가까울 수록 좋음)
# (2) 따라서 Residual은 정규분포 특성을 가져야한다. (정규분포 모양에 가까울 수록 좋음)
# (3) 정규분포 특성을 조사하면  Fitting이 잘되었는지 확인할 수 있다. : Q-Q plot
# (4) 잔차 성분의 ACF : 자기상관관계가 작을 수록 좋다.
residual = model_fit.resid
checkNormality(residual)  # 육안으로 백색 잡음 형태인지 확인한다
```

# ARIMA 모형 : ARIMA(1,1,1) 모형의 시계열 예측
- a = 0.8, b=0.5, 차분횟수 = 1 인 ARIMA(1,1,1) 샘플 데이터를 생성한 후 향후 10 기간 동안의 예측값을 추정한다 . 예측 결과가 의미하는 것은 무엇인가

```python
# 향후 10 기간 데이터를 예측한다
forecast = model_fit.forecast(steps=10)[0]
forecast = np.r_[y[-1], forecast]  # y의 마지막 값을 forecast 앞 부분에 넣는다

# 원 시계열과 예측된 시계열을 그린다
ytail = y[len(y)-100:]   # 뒷 부분 100개만 그린다
ax1 = np.arange(1, len(ytail) + 1)
ax2 = np.arange(len(ytail), len(ytail) + len(forecast))
plt.figure(figsize=(10, 6))
plt.plot(ax1, ytail, 'b-o', markersize=3, color='blue', label='Time series', linewidth=1)
plt.plot(ax2, forecast, color='red', label='Forecast')
plt.axvline(x=ax1[-1],  linestyle='dashed', linewidth=1)
plt.title("Time Series Forcast")
plt.legend()
plt.show()
```

# ARIMA 모형 : 주가 예측
- 주가 데이터에 ARIMA 모형을 적용하고 ARIMA 모형으로 주가를 설명할 수 있는지 확인한다. -> 주가 데이터는 ARIMA 모형으로 설명하기 어렵다 . 그 이유는?

```python
p = pd.read_csv('StockData/102110.csv', index_col=0, parse_dates=True)

# 종가를 기준으로 일일 수익률을 계산한다.
p['Rtn'] = np.log(p['Close']) - np.log(p['Close'].shift(1))
p = p.dropna()

# 수익률 시계열을 육안으로 확인한다. 이분산성이 있는가?
plt.figure(figsize=(15,10))
plt.plot(p['Rtn'], color='red', linewidth=1)
plt.show()

# 주가 데이터를 ARIMA(2,1,1) 모형으로 분석한다 (Fitting)
y = np.array(pd.to_numeric(p['Close'], downcast='float'))  # int형이면 float형으로 변환한다
model = ARIMA(y, order=(1,1,1))
model_fit = model.fit()
print(model_fit.summary())

# Fitting이 잘되었는지 확인하기 위해 Residual을 분석한다.
# (1) Residual은 실제 데이터와 추정치의 차이이므로 백색 잡음 (잔차) 이어야 한다. (백색 잡음에 가까울 수록 좋음)
# (2) 따라서 Residual은 정규분포 특성을 가져야한다. (정규분포 모양에 가까울 수록 좋음)
# (3) 정규분포 특성을 조사하면  Fitting이 잘되었는지 확인할 수 있다. : Q-Q plot
# (4) 잔차 성분의 ACF : 자기상관관계가 작을 수록 좋다.
residual = model_fit.resid
checkNormality(residual)  # 육안으로 백색 잡음 형태인지 확인한다
```
