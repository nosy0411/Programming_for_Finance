```python
import numpy as np
import matplotlib.pyplot as plt
# 주기가 8인 코사인 그래프로 시작해보자
t = np.arange(32)

f=1.0/8

y=np.cos(2*np.pi*f*t)

plt.plot(y)
plt.show()

print(t.shape[-1])
```

```python
# numpy로 fft 분석을 해보자
sp = np.fft.fft(y)

freq = np.fft.fftfreq(t.shape[-1])

print(sp)
print(freq)

plt.plot(freq, sp)
plt.show()
```

```python
# 이 주파수 스펙트럼을 역푸리에 변환하면 다시 원래의 모습으로 돌아온다.
iy = np.fft.ifft(sp)

print(iy)
plt.plot(iy)
plt.show()
```

```python
# 좀 더 다양한 주파수와 함수에 적용시켜 보자
f1 = 2  # 주파수 입력
f2 = 3
f3 = 5
f4 = 21
f5 = 30

w1 = 2*np.pi*f1  # w = 2pi*f
w2 = 2*np.pi*f2
w3 = 2*np.pi*f3
w4 = 2*np.pi*f4
w5 = 2*np.pi*f5

print("f1 =",f1, "/ f2 =",f2, "/ f3 =",f3, "/ f4 =",f4, "/ f5 =",f5 )
print("w1 =",w1, "/ w2 =",w2, "/ w3 =",w3, "/ w4 =",w4, "/ w5 =",w5 )

t = np.arange(0,10,0.01)   # sampling freq : 100 Hz // sampling time : 10 seconds
#t = np.arange(0:10:0.001)  # sampling freq : 1000 Hz
x = np.sin(w1*t) + 3*np.sin(w2*t) + 7*np.sin(w3*t) + 21*np.sin(w4*t) + 30*np.sin(w5*t)

print(t.shape)

plt.plot(t, x, 'r-')
plt.xlabel('t',fontsize=20)
plt.ylabel('x',fontsize=20)
#plt.xlim(0,1)
plt.show()
```

```python
# Sampling frequency 계산 : Fs = 데이터 개수/시간
import math

Fs = int(len(t)/t[-1])
print(Fs)

# FFT 연산
NFFT   =  len(x)                           # length of signal > NFFT = 2 ^ nextpow2(length(y)) 
print(NFFT)

k      =  np.arange(NFFT)
print(k)

f0     =  k * Fs / NFFT                    # Double sides Frequency range
# print(f0)
f0     =  f0[range(math.trunc(NFFT/2))]    # Single side  Frequency range
# print(f0)

Y      =  np.fft.fft(x)/NFFT               # FFT computing and normalization
Y      =  Y[range(math.trunc(NFFT/2))]     # Single side  Frequency range
print(Y)

amp_Hz =  2 * abs(Y)
print(amp_Hz)
phase_ang = np.angle(Y) * 180/np.pi
print(phase_ang)
```

```python
plt.figure(figsize=(16,5))
plt.plot(f0,amp_Hz,'r.-', lw=1)
plt.xlim(0, Fs/2)
plt.ylim(0)
plt.xticks(np.arange(0,(Fs/2)+1))
plt.xlabel('frequency(Hz)')
plt.ylabel('amplitude')
plt.grid()
```

```python
# 이 원리를 이용해서 주가를 푸리에 변환해보고 주파수를 일정범위만 적용해보면 이평선처럼 상당히 평활화된 결과를 확인해볼 수 있다.

import FinanceDataReader as fdr

df = fdr.DataReader('035720', '2010-01-01')

stock_price = df['Close']

sp = np.fft.fft(stock_price)
freq = np.fft.fftfreq(len(stock_price))


print(sp)
print(freq)

plt.figure(figsize=(15,8))
plt.plot(freq, sp)
plt.show()
```

```python
csp = np.copy(sp)
print(csp)

# 특정 주기. 즉 주파수를 가지는 것은 제외한다. 3:-3 인덱스 범위의 주파수가 그렇게 크게 나타나지 않으므로 제거한다.
csp[3:-3]=0

# 가격 주파수 스펙트럼을 역푸리에 변환한다. 평활화된 주가가 나오게 된다.
iffted_stock_price = np.fft.ifft(csp)

print(iffted_stock_price)
print(len(stock_price))
print(len(iffted_stock_price))

import pandas as pd
# 역푸리에 변환한 주가에 날짜를 대입해준다.
ifft_df = pd.DataFrame(iffted_stock_price, index=stock_price.index)

# 두 주가를 시각화해준다. 
plt.figure(figsize=(15,8))
plt.plot(stock_price)
plt.plot(ifft_df)
plt.show()
```
