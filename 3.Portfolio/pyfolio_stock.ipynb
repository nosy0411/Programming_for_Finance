## 백테스팅 라이브러리 종류

- backtest
    
    - 여러가지 트레이딩 전략을 구현하는데 있어 매우 쉽고 직관적인 인터페이스 제공, 백테스팅 결과물을 상호작용 방식의 시각화 툴을 사용해 예쁘게 표시해준다. 
    
    - 기술적 지표 라이브러리 talib과 호환성이 매우 좋다

    - 주소 : https://kernc.github.io/backtesting.py/

- bt 

    - 쉽게 검증이 가능하면서 재사용이 가능한 전략 로직의 블럭들을 개발하여 아주 복잡한 트레이딩 전략의 빠른 구현이 가능하다. 
    
    - 특히나 포트폴리오 기반의 퀀트 전략들을 검증해보는데 초점이 맞춰져 있으며 자산 비중 배분이나 포트폴리오 리밸런싱에 대한 알고리즘을 장착하고 있다. 
    
    - 또한 코드를 새로 짤 필요가 없이 조금만 변화를 주면 동일 전략을 다른 시간 혹은 다른 자산 비중으로 백테스팅하는 것도 어렵지 않게 설계되어 있다.

    - 주소 : https://pmorissette.github.io/bt/

- backtrader

    - 다른 라이브러리에 비해 사용법에 대한 메뉴얼이 잘 구성되어 있다. 그리고 해당 라이브러리를 사용하는 사람들의 커뮤니티가 잘 구성되어 있다.

    - csv 형태나 pandas의 데이터프레임 형태를 포함한 여러 가지 데이터 형식들을 처리할 수 있으며, 해외 온라인 브로커리지 회사를 통한 실시간 데이터 수신도 가능하다

    - 주소 : https://www.backtrader.com/

- zipline

    - IPython Notebook 인터페이스로 접근이 가능한 이 라이브러리는 상호작용 방식의 커맨드라인 인터페이스 덕분에 사용이 매우 편리하다.

    - Quantopian이라는 퀀트 트레이딩 회사에 의해 개발되었으며, 전략 개발부터 백테스팅, 그리고 주문 집행까지 통합적인 백테스팅 프레임워크를 제공하고 있다.

    - 또한 방대한 양의 미국 주식 데이터를 제공하는 알고리즘 트레이딩 시뮬레이터이다.

    - 주소 : https://github.com/quantopian/zipline
