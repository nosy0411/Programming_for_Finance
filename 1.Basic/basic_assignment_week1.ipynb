## 과제 4 문제
![과제 4](./week1_4.png)

```python
# 과제4 해답 1

def frequency(string):
    string = string.lower()
    dict = {}
    for n in string:
        keys = dict.keys()
        if n == ' ':
            continue 
        elif n in keys:
            dict[n] += 1       
        else:
            dict[n] = 1
    return dict
print(frequency('@#$@#$@e'))
```

```python
# 과제 4 해답 2

def countAllLetters(string):
    number = dict()
    for i in string:
        if str.isalpha(i): #이게 다른 특수기호나 문자가 아닌 letter만을 count해준다고 함.
            i = i.upper() #대문자 소문자 둘다 같은 것으로 고려해주는 것
            if i not in number:
                number[i] = 1
            else:
                number[i] += 1
    return number

a = input('Enter the string: ')
print(countAllLetters(a))
```

## 과제 6 문제

```python
# 과제 6 해답 1

def count_func(characters, finder):
    if finder in characters:
        count = len(characters.split(finder))-1
    else:
        count = 0
    return count

# 과제 6 해답 2
def count_func(s,item):
    count=0
    i=0
    while(1):
        if i>=(len(s) - len(item) + 1):
            break
        elif s[i:i+len(item)]==item:
            count+=1
            i=i+len(item)
        else:
            i=i+1
    return count

print(count_func('representative','re'))
print(count_func('aaaaaabbaa','aa'))
print(count_func('aabaabbaabaaabaabbbaabbbbba','aa'))
print(count_func('aaaaaabbaa','aa'))

# 과제 6 해답 3
def count_fuc(string, substring): 
    x = 0
    start = 0
    while True: 
        a = string.find(substring, start) 
        if a > -1: 
            x += 1
            start = a + len(substring)    
        else: 
            break
    return x 
print(count_fuc('aaaaaabbaa', 'aa'))
```
